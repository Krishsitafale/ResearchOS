```python
import torch
import torch.nn as nn
import torch.optim as optim

# 1. Create fake data following the secret rule: Output = (2 * X1) + (3 * X2) + 5
# We create 100 rows of data, each with 2 input numbers
X = torch.randn(100, 2)
y = (2 * X[:, 0:1]) + (3 * X[:, 1:2]) + 5

# 2. Define the Model (Your Linear Layer)
model = nn.Linear(2, 1)

# 3. Define the Loss Function (Criterion) and the Optimizer
criterion = nn.MSELoss()  # Mean Squared Error: Measures how far off the guesses are
optimizer = optim.SGD(model.parameters(), lr=0.1) # Stochastic Gradient Descent with learning rate

# 4. Check the random weights BEFORE training
print("--- BEFORE TRAINING ---")
print("Weights:", model.weight.data)
print("Bias:", model.bias.data)
print("\nTraining the model now...\n")

# 5. The Training Loop (100 Epochs)
epochs = 100
for epoch in range(epochs):
    # Step 1: Make a guess
    prediction = model(X)

    # Step 2: Check the answer key (Calculate loss)
    loss = criterion(prediction, y)

    # Step 3: Erase old notes
    optimizer.zero_grad()

    # Step 4: Find out who to blame (Backward pass)
    loss.backward()

    # Step 5: Fix the mistakes (Update weights)
    optimizer.step()

    # Print progress every 20 rounds
    if (epoch + 1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# 6. Check the learned weights AFTER training
print("\n--- AFTER TRAINING ---")
print("Weights (Target is [[2.0, 3.0]]):", model.weight.data)
print("Bias (Target is [5.0]):", model.bias.data)
```
