    Dataset
      │
      ▼
    Prediction
      │
      ▼
    Loss
      │
      ▼
    Gradient
      │
      ▼
    Update Weight
      │
      ▼
    Repeat

```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
X = np.array([1,2,3,4,5], dtype=float)
y = np.array([30,42,55,67,79], dtype=float)

# Parameters
w = 1.0
b = 0.0

learning_rate = 0.01
epochs = 500

loss_history = []

n = len(X)

for epoch in range(epochs):

    # Prediction
    y_pred = w * X + b

    # Loss
    loss = np.mean((y - y_pred) ** 2)
    loss_history.append(loss)

    # Gradients
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # Update
    w -= learning_rate * dw
    b -= learning_rate * db

print(f"Weight: {w:.4f}")
print(f"Bias: {b:.4f}")
print(f"Final Loss: {loss:.4f}")
```

```python
plt.figure(figsize=(8,5))

plt.plot(loss_history)

plt.title("Loss vs Epoch")

plt.xlabel("Epoch")

plt.ylabel("Loss")

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.scatter(X, y, label="Actual Data")

plt.plot(X, w*X+b, color="red", label="Learned Line")

plt.legend()

plt.show()
```
