# Numerical Gradient From Scratch

## Objective

Understand how a machine can determine which direction reduces the loss function without using calculus.

---

## Learning Outcomes

- Understand numerical differentiation
- Estimate gradients
- Visualize loss
- Prepare for Gradient Descent

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5], dtype=float)

y = np.array([30, 42, 55, 67, 79], dtype=float)
```

```python
plt.figure(figsize=(6,4))

plt.scatter(X, y)

plt.xlabel("Hours Studied")

plt.ylabel("Marks")

plt.title("Training Dataset")

plt.grid(True)

plt.show()
```

```python
def predict(X, w, b):
    return w * X + b
```

```python
w = 1.0
b = 0.0
```

```python
def mse(y_true, y_pred):

    return np.mean((y_true - y_pred) ** 2)
```

```python
prediction = predict(X, w, b)

print(prediction)
```

```python
current_loss = mse(y, prediction)

print(current_loss)
```

```python
epsilon = 0.0001
new_w = w + epsilon
```

```python
new_prediction = predict(X, new_w, b)
```

```python
new_loss = mse(y, new_prediction)

print(new_loss)
```

```python
gradiant = (new_loss - current_loss) / epsilon
print(gradiant)
```

```python
weights = [1, 3, 5, 7, 9, 11]

for w in weights:

    prediction = predict(X, w, b)

    loss = mse(y, prediction)

    print(f"Weight = {w}")

    print(f"Loss = {loss:.2f}")

    print()
```

```python
weights = np.linspace(0, 20, 100)

losses = []

for w in weights:

    prediction = predict(X, w, b)

    losses.append(mse(y, prediction))
```

```python
plt.figure(figsize=(8,5))

plt.plot(weights, losses)

plt.xlabel("Weight")

plt.ylabel("Loss")

plt.title("Loss Landscape")

plt.grid(True)

plt.show()
```
