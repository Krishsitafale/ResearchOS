```python
plt.figure(figsize=(8,2))

words = ["The","rover","collected","soil","on","Mars"]

x = np.arange(len(words))

plt.scatter(x,[0]*len(words),s=300)

for i,w in enumerate(words):
    plt.text(i,-0.1,w,ha='center')

plt.plot([1,3],[0,0])
plt.plot([3,5],[0,0])

plt.axis("off")

plt.show()
```

Self Attention

Each word asks
Which other words help me understand my meaning?
Example
soil

↓

rover

↓

Mars

↓

collected
Not every word is equally important.

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)

sentence = "Autonomous rover navigation"

embedding = model.encode(sentence)

print(embedding.shape)
```
