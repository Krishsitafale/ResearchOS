```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
sentences = [

    "ROS2 controls autonomous robots.",

    "Robot Operating System is used for autonomous navigation.",

    "Machine Learning improves perception.",

    "Deep Learning enables computer vision.",

    "The weather is pleasant today.",

    "A Mars rover navigates rocky terrain.",

    "Autonomous vehicles require localization."
]
```

```python
embeddings = model.encode(sentences)
```

```python
print(embeddings.shape)
```

```python
print(embeddings[0])
```

```python
similarity = cosine_similarity([embeddings[0]], [embeddings[1]])
print(similarity)
```

```python
similarity = cosine_similarity(
    [embeddings[0]],
    [embeddings[4]]
)

print(similarity)
```

```python
similarity = cosine_similarity(
    [embeddings[4]],
    [embeddings[5]]
)

print(similarity)
```

```python
for i in range(len(sentences)):
  for j in range(i+1, len(sentences)):

    score = cosine_similarity(
        [embeddings[i]],
        [embeddings[j]]
    )[0][0]

    print(f"{i} <-> {j} : {score:.3f}")
```
